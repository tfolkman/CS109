```python
%matplotlib inline
from urllib import urlopen

import brewer2mpl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set up some better defaults for matplotlib
from matplotlib import rcParams

#colorbrewer2 Dark2 qualitative color table
dark2_colors = brewer2mpl.get_map('Dark2', 'Qualitative', 7).mpl_colors

rcParams['figure.figsize'] = (10, 6)
rcParams['figure.dpi'] = 150
rcParams['axes.color_cycle'] = dark2_colors
rcParams['lines.linewidth'] = 2
rcParams['axes.facecolor'] = 'white'
rcParams['font.size'] = 14
rcParams['patch.edgecolor'] = 'white'
rcParams['patch.facecolor'] = dark2_colors[0]
rcParams['font.family'] = 'StixGeneral'


def remove_border(ax1=None, axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecesasry plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    
    if ax1 == None:
        ax = axes or plt.gca()
    else:
        ax = ax1
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
```

### Bar Chart

```python
fig, ax = plt.subplots(figsize=(3,8))

change = grad.change[grad.change > 0]
city = grad.city[grad.change > 0]
pos = np.arange(len(change))

ax.set_title('1995-2005 Change in HS graduation rate')
ax.barh(pos, change)

#add the numbers to the side of each bar
for p, c, ch in zip(pos, city, change):
    ax.annotate(str(ch), xy=(ch + 1, p + .5), va='center')
    
#cutomize ticks
ax.set_yticks(pos + .5)
ax.set_yticklabels(city)
plt.setp(ax.get_xticklabels(), visible=False)

#minimize chartjunk
remove_border(left=False, bottom=False)
ax.grid(axis = 'x', color ='white', linestyle='-')

#set plot limits
ax.set_ylim(pos.max(), pos.min())
ax.set_xlim(0, 30)
```

```python
years = np.arange(2004, 2009)
heights = np.random.random(years.shape) * 7000 + 3000

boxcolors = brewer2mpl.get_map('Set1', 'qualitative', 5).mpl_colors

fig, ax = plt.subplots()

ax.bar(years, heights, color=boxcolors, align = 'center')
ax.grid(axis='y', color='white', ls='-', lw=1)
ax.set_yticks([2000, 4000, 6000, 8000])

fmt = plt.ScalarFormatter(useOffset=False)
plt.gca().xaxis.set_major_formatter(fmt)
ax.set_xlim(2003.5, 2008.5)
remove_border(left=False)

for x, y in zip(years, heights):
    ax.annotate("{:,.0f}".format(y), (x, y + 200), ha='center')
```

### Scatter Plot

```python
fig, axes = plt.subplots(1, 2, figsize=(6, 4))

axes[0].scatter(diamonds.carat, diamonds.price, color="k")
axes[0].set_ylim(0, diamonds.price.max())
axes[0].set_xlim(0, 5)
axes[0].set_xlabel('Carat')
axes[0].set_ylabel('Price')
remove_border(ax1=axes[0])

axes[1].scatter(diamonds.carat, diamonds.price, color='k', alpha=.01)
axes[1].set_ylim(0, diamonds.price.max())
axes[1].set_xlim(0, 5)
axes[1].set_xlabel('Carat')
axes[1].set_ylabel('Price')
remove_border(ax1=axes[1])

fig.tight_layout()
```

### Trend Line

```python
x = diamonds.carat[diamonds.carat < 2]
y = diamonds.price[diamonds.carat < 2]

fig, ax = plt.subplots()

ax.plot(x, y, 'o', mec='none', alpha=.05)

#fit and overplot 2nd order polynomial
params = np.polyfit(x, y, 2)
xp = np.linspace(x.min(), 2, 20)
yp = np.polyval(params, xp)
ax.plot(xp, yp, 'k')

#overplot an error band
sig = np.std(y - np.polyval(params, x))
plt.fill_between(xp, yp - sig, yp + sig, 
                 color='k', alpha=0.2)

plt.xlabel("Carat")
plt.ylabel("Price")
plt.xlim(0, 2)
remove_border()
```
